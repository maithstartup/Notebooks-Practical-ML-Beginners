Seaborn Versus Matplotlib.

Simple random-walk plot in Matplotlib, using its classic plot formatting and colors. We start with the typical imports:

```
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
```

create some random walk data

```
# Create some data
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)
```

```
# Plot the data with Matplotlib defaults
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

The result contains all the information to convey.

Now let's take a look at how it works with Seaborn.  Seaborn has many of its own high-level plotting routines, but it can also overwrite Matplotlib's default parameters and in turn get even simple Matplotlib scripts to produce vastly superior output.

We can set the style by calling Seaborn's set() method. By convention, Seaborn is imported as sns

```
import seaborn as sns
sns.set()
```

Now let's rerun the same two lines as before

```
# same plotting code as above!
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

```
sns.set(style="white")
```

```
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
```
