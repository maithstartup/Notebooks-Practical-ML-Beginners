```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

```
sns.get_dataset_names()
```

```
data =sns.load_dataset("diamonds")
```

```
data.head()
```

```
data.info()
```

```
np.random.seed(101)
data_min=data.sample(500)
```

```
data_min
```

# Scatter Plot

```
sns.relplot(x="carat",y="price",data=data_min)
```

```
sns.relplot(x="carat",y="price",hue="cut",data=data_min)
```

```
sns.scatterplot(x="carat",y="price",hue="color",data=data_min)
```

```
sns.scatterplot(x="depth",y="price",data=data_min)
```

```
sns.scatterplot(x="carat",y="price",size="table",data=data_min)
```

```
sns.relplot(x="carat",y="price",col="cut",col_wrap=3, data=data_min)
```

```
data_min.corr()
```

# Line Plot

```
np.random.seed(105)
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)

g.fig.autofmt_xdate()
```

```
data_2=sns.load_dataset("fmri")
```

```
data_2.head()
```

```
data_2.info()
```

```
sns.lineplot(x="timepoint",y="signal",data=data_2)
```

```
sns.lineplot(x="timepoint",y="signal",ci="sd",data=data_2)
```

```
sns.lineplot(x="timepoint", y="signal", ci=None, data=data_2);
```

```
sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=data_2);
```

```
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            kind="line", data=data_2);
```

```
df = pd.DataFrame(dict(time=pd.date_range("2017-1-1", periods=500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
```

# Multiple Plots

```
sns.relplot(x="timepoint", y="signal",
            col="region", height=3,
            kind="line", data=data_2);
```

```
sns.relplot(x="timepoint", y="signal",
            col="region", row="event", height=5,
            kind="line", data=data_2);
```

```
sns.relplot(x="timepoint", y="signal", hue="event", style="event",
            col="subject", col_wrap=5,
            height=3, aspect=.75, linewidth=2.5,
            kind="line", data=data_2.query("region == 'frontal'"));
```
