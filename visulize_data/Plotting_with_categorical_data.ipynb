Import the libraries

```
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

Seaborn offers some built-in data sets.


```
sns.get_dataset_names()
```

Import the datasets.

Titanic Datasets- data frames describe the survival status of individual passengers on the Titanic. 

iris Dataset:-Iris flower data set.The data set consists of 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor).

Tips dataset-Food serversâ€™ tips in restaurants may be influenced by many
factors, including the nature of the restaurant, size of the party, and table locations in the restaurant.

Diamonds Dataset-classic dataset contains the prices and other attributes.


```
titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
tips = sns.load_dataset("tips")
diamonds = sns.load_dataset("diamonds")
```

Categorical Scatter Plots

stripplot()-is used when one of the variable under study is categorical. It represents the data in sorted order along any one of the axis.

```
sns.stripplot(x="species", y="petal_length", data=iris)
plt.show()
```

Categorical scatter plots with non overlapping points

```
sns.swarmplot(x="species", y="petal_length", data=iris)
plt.show()
```

```
sns.catplot(x="smoker", y="tip", order=["No", "Yes"], data=tips,jitter=False);
```

Distribution plots

As the size of the dataset grows, categorical scatter plots become limited in the information they can provide about the distribution of values within each category. When this happens, there are several approaches for summarizing the distributional information in ways that facilitate easy comparisons across the category levels.

Boxplot-This kind of plot shows the three quartile values of the distribution along with extreme values. Each value in the boxplot corresponds to an actual observation in the data.

```
sns.boxplot(x="alive",y="age",hue="adult_male",data=titanic)
```

```
sns.boxplot(x="day", y="total_bill",data=tips);
```

Box plot with wide form format

```
sns.boxplot(data=iris,orient="h")
```

Violin plot-A different approach is a violinplot(), which combines a boxplot with the kernel density estimation(KDE).

KDE-useful for plotting the shape of a distribution.

```
sns.violinplot(x="age", y="sex",hue="survived",data=titanic )
```

```
sns.violinplot(x="age", y="sex",hue="survived",inner=None ,data=titanic )
```

boxen plot-draws a plot that is similar to a box plot but optimized for showing more information about the shape of the distribution. It is best suited for larger datasets.

```
sns.boxenplot(x="color", y="price", data=diamonds.sort_values("color") );
```

Categorical Estimate plots

Bar plot-show point estimates and confidence intervals.
In seaborn, the barplot() function operates on a full dataset and applies a function to obtain the estimate (taking the mean by default).

```
sns.barplot(x="sex", y="survived", hue="class",data=titanic)
```

count plot-shows the count of observations

```
sns.countplot(x="deck",data=titanic,palette="Greens_d")
```

point plot-This function also encodes the value of the estimate with height on the other axis, but rather than showing a full bar, it plots the point estimate and confidence interval.


```
sns.pointplot(x="sex", y="survived", hue="class", kind="point", data=titanic);
```

While the categorical functions lack the style semantic of the relational functions, it can still be a good idea to vary the marker and/or linestyle along with the hue to make figures that are maximally accessible and reproduce well in black and white.

```
sns.pointplot(x="class", 
y="survived", 
hue="sex", 
data=titanic,
palette={"male":"g",
"female":"m"},
markers=["^","o"],
linestyles=["-","--"])
```

```
sns.catplot(x="day", y="total_bill", hue="smoker",
            col="time", aspect=.6,
            kind="swarm", data=tips);
```

```
sns.catplot(x="day", y="total_bill", hue="time",
            col="smoker", aspect=.6,
            kind="swarm", data=tips);
```
